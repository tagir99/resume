```python
from scipy import stats

t = -2.5
n = 39
df = n - 1

p = 2 * stats.t.sf(abs(t), df)
print(f'p = {p:.3f}')

''' Далее, чем 2.5 сигмы в т - распределении с 38 н - значений находятся 1,7% значений '''
```
